```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = 'https://fbref.com/en/comps/Big5/Big-5-European-Leagues-Stats'

response = requests.get(url)
soup = BeautifulSoup(response.text)
big_five_league = soup.select('table.stats_table')[0]
links = big_five_league.find_all('a')
```

```python
links = [l.get('href') for l in links]
links = [l for l in links if '/squads/' in l]
links = ['https://fbref.com' + l for l in links]
links
```

```python
links_with_2023 = [l for l in links if '2023' in l]
links_without_2023 = [l for l in links if'2023' not in l]
```

```python
links_without_2023
```

```python
codes_teams = {}
split_links_with_2023 = [(url.split('/')[5], url.split('/')[7]) for url in links_with_2023]

for code, team_name in split_links_with_2023:
    codes_teams[team_name] = code
    print(f"Code : {code}   Team Name : {team_name}")
```

```python
split_links_without_2023 = [(url.split('/')[5], url.split('/')[6]) for url in links_without_2023]

for code, team_name in split_links_without_2023:
    codes_teams[team_name] = code
    print(f"Code : {code}   Team Name : {team_name}")
```

```python
codes_teams
```

```python
data_dict = {}
for key, value in codes_teams.items():
    url = 'https://fbref.com/en/squads/' + value + '/2023-2024/all_comps/' + key + '-All-Competitions'
    data = requests.get(url)
    content = pd.read_html(data.text, match = 'Standard Stats')
    data_dict[key] = content
    time.sleep(30)
```

```python
df_list = []

for team in data_dict.keys():
    df = pd.DataFrame(data_dict[team][0])
    df.columns = df.columns.droplevel(0)
    print(f"Team : {team}   Col count : {len(df.columns)}")
    df['Team'] = team
    df['Team'] = df['Team'].str.rstrip('-Stats')
    df_list.append(df)

final_df = pd.concat(df_list)

final_df
```

```python
cols_to_keep = ['Player', 'Nation', 'Pos', 'Age', 'MP', 'Starts', 'Min', '90s', 'Gls',
       'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 'Gls', 'Ast',
       'G+A', 'G-PK', 'G+A-PK', 'Matches', 'Team']
cols_to_drop = set(final_df.columns.tolist()) - set(cols_to_keep)
cols_to_drop
```

```python
final_df.to_csv('big_five_leagues_with_additional_cols.csv', index=False) # ekstra collar işe yarar mı kontrol etmek için
```

```python
df_super_lig = pd.read_csv("player_data_super_league.csv")
cols_to_drop = [tup for tup in cols_to_drop]

big_five_revised = final_df.drop(columns=cols_to_drop, errors='ignore')
big_five_revised.columns = ['Player', 'Nation', 'Pos', 'Age', 'MP', 'Starts', 'Min', '90s', 'Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 'Gls_per_90', 'Ast_per_90', 'G+A_per_90', 'G-PK_per_90', 'G+A-PK_per_90', 'Matches','Team']
```

```python
big_five_revised['Nation'] = big_five_revised['Nation'].str[-3:]
big_five_revised.fillna(value = 'Unknown',inplace=True)
```
