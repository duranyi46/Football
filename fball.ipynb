```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
urls_with_2023 = ['https://fbref.com/en/squads/ecd11ca2/2023-2024/Galatasaray-Stats',
 'https://fbref.com/en/squads/ae1e2d7d/2023-2024/Fenerbahce-Stats',
 'https://fbref.com/en/squads/c8ad3091/2023-2024/Trabzonspor-Stats',
 'https://fbref.com/en/squads/bff39cf5/2023-2024/Istanbul-Basaksehir-Stats',
 'https://fbref.com/en/squads/d6ffefd1/2023-2024/Kasimpasa-Stats',
 'https://fbref.com/en/squads/0f9294bd/2023-2024/Besiktas-Stats',
 'https://fbref.com/en/squads/c63f6c0c/2023-2024/Sivasspor-Stats',
 'https://fbref.com/en/squads/076128d7/2023-2024/Alanyaspor-Stats',
 'https://fbref.com/en/squads/f622e63c/2023-2024/Rizespor-Stats',
 'https://fbref.com/en/squads/5ac76942/2023-2024/Antalyaspor-Stats',
 'https://fbref.com/en/squads/7c81865f/2023-2024/Gaziantep-FK-Stats',
 'https://fbref.com/en/squads/d143cf0a/2023-2024/Adana-Demirspor-Stats',
 'https://fbref.com/en/squads/77356e9e/2023-2024/Samsunspor-Stats',
 'https://fbref.com/en/squads/1f33fbc7/2023-2024/Kayserispor-Stats',
 'https://fbref.com/en/squads/4b5ddee5/2023-2024/Hatayspor-Stats',
 'https://fbref.com/en/squads/5700c020/2023-2024/Konyaspor-Stats']

urls_without_2023 = ['https://fbref.com/en/squads/b88463bd/Ankaragucu-Stats',
 'https://fbref.com/en/squads/bd61c555/Fatih-Karagumruk-Stats',
 'https://fbref.com/en/squads/2f116859/Pendikspor-Stats',
 'https://fbref.com/en/squads/6b2d5f93/Istanbulspor-Stats']
codes_teams = {}
```

```python
split_data_with_2023 = [(url.split('/')[5], url.split('/')[7]) for url in urls_with_2023]


for code, team_name in split_data_with_2023:
    codes_teams[team_name] = code
    print(f"Code: {code}, Team Name: {team_name}")
```

```python
split_data_without_2023 = [(url.split('/')[5], url.split('/')[6]) for url in urls_without_2023]

for code, team_name in split_data_without_2023:
    codes_teams[team_name] = code
    print(f"Code: {code}, Team Name: {team_name}")
```

```python
data_dict = {}
for key, value in codes_teams.items():
    url = 'https://fbref.com/en/squads/' + value + '/2023-2024/all_comps/' + key + '-All-Competitions'
    data = requests.get(url)
    content = pd.read_html(data.text, match = 'Standard Stats')
    data_dict[key] = content
    time.sleep(30)
```

```python
df1 = pd.DataFrame(data_dict['Galatasaray-Stats'][0])
df2 = pd.DataFrame(data_dict['Konyaspor-Stats'][0])
cols_df1 = set(df1.columns)
cols_df2 = set(df2.columns)
diff = cols_df1 - cols_df2
remove = ('Unnamed: 33_level_0', 'Matches')
diff.discard(remove)
cols_to_drop = [tup[1] for tup in diff]

print(diff)
```

```python
# List to store individual DataFrames
df_list = []

# Process each key in the dictionary
for team in data_dict.keys():
    # Extract the DataFrame from the dictionary
    df = pd.DataFrame(data_dict[team][0])
    
    # Drop the first level of the MultiIndex columns
    df.columns = df.columns.droplevel(0)
    
    # Drop columns that are in cols_to_drop if they exist
    cols_to_drop_existing = [col for col in cols_to_drop if col in df.columns]
    df = df.drop(columns=cols_to_drop_existing, errors='ignore')
    
    # Check the number of columns
    print(f"{team} columns: {df.columns}")
    
    # Rename the columns if the number of columns matches the expected number
    if len(df.columns) == 22:
        df.columns = ['Player', 'Nation', 'Pos', 'Age', 'MP', 'Starts', 'Min', '90s', 'Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 'Gls_per_90', 'Ast_per_90', 'G+A_per_90', 'G-PK_per_90', 'G+A-PK_per_90', 'Matches']
    else:
        print(f"Unexpected number of columns for {team}: {len(df.columns)}")
    
    # Add a column to identify the team
    df['Team'] = team
    
    # Append the DataFrame to the list
    df_list.append(df)

# Concatenate all DataFrames into a single DataFrame
final_df = pd.concat(df_list, ignore_index=True)

# Display the final DataFrame
final_df
```

```python
final_df['Team'] = final_df['Team'].str.rstrip('-Stats')
final_df['Nation'] = final_df['Nation'].str[-3:]
final_df.fillna(value = 'Unknown', inplace = True)
final_df
```

```python
final_df.to_csv('player_data_super_league.csv', index=False)
```
