```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
```

```python
url1 = 'https://fbref.com/en/comps/26/2023-2024/2023-2024-Super-Lig-Stats'
response1 = requests.get(url1)
soup = BeautifulSoup(response1.text)
super_league_teams = soup.select('table.stats_table')[0]
spl_links = super_league_teams.find_all('a')
spl_links = [l.get('href') for l in spl_links]
spl_links = [l for l in spl_links if '/squads/' in l]
spl_links = ['https://fbref.com' + l for l in spl_links]
```

```python
url2 = 'https://fbref.com/en/comps/Big5/Big-5-European-Leagues-Stats'
response2 = requests.get(url2)
soup2 = BeautifulSoup(response2.text)
five_big_league_teams = soup2.select('table.stats_table')[0]
vbl_links = five_big_league_teams.find_all('a')
vbl_links = [l.get('href') for l in vbl_links]
vbl_links = [l for l in vbl_links if '/squads/' in l]
vbl_links = ['https://fbref.com' + l for l in vbl_links]
```

```python
all_team_links = spl_links + vbl_links
```

```python
all_team_links
```

```python
matchlog_links = []
for link in all_team_links:
    url = link  
    try:
        resp = requests.get(url)
        soup = BeautifulSoup(resp.text)  
        
        content = soup.select('table.stats_table')[0]
        links = content.find_all('a')
        links = [l.get('href') for l in links]
        links = [l for l in links if '/matchlogs/' in l]
        links = ['https://fbref.com' + l for l in links]
        
        if links:
            matchlog_links.append(links)
            print('success')
        else:
            matchlog_links.append('Link is not available')
            print('couldnt find but success')
        
    except Exception as e:
        matchlog_links.append('Link is not available')
        print('couldnt find but success')
    
    time.sleep(20)
    
```

```python
# Flatten the list of lists
flattened_list = [url for sublist in matchlog_links for url in sublist]


len(flattened_list)
```

```python
# Create the dictionary
players_dict = {url.split('/')[-1].replace('-Match-Logs', '').replace('-', ' '): url for url in flattened_list}
```

```python
dataset = pd.read_csv("big_five_plus_super_league.csv")
dataset
```

```python
players_df = pd.DataFrame(list(players_dict.items()), columns=['Player', 'url'])
len(players_df)
```

```python
import pandas as pd
import unidecode

def normalize_column(df, column):
    df[column + '_norm'] = df[column].apply(lambda x: unidecode.unidecode(x).lower())
    return df

# Normalize the 'Player' column in both DataFrames
dataset = normalize_column(dataset, 'Player')
players_df = normalize_column(players_df, 'Player')

# Merge the DataFrames using the normalized column
merged_df = pd.merge(dataset, players_df, left_on='Player_norm', right_on='Player_norm', how='left')

# Drop the temporary normalized column
merged_df.drop(['Player_norm'], axis=1, inplace=True)

# Optionally, if you want to keep the original case of 'Player' from dataset
merged_df['Player'] = dataset['Player']
```

```python
# Merge the DataFrames using the normalized column
merged_df = pd.merge(dataset, players_df, left_on='Player_norm', right_on='Player_norm', how='left')

# Drop the temporary normalized column
merged_df.drop(columns=['Player_norm','Player_y'], axis=1, inplace=True)
merged_df.rename(columns = {'Player_x':'Player'}, inplace = True)
```

```python
merged_df.rename(columns = {'url':'Matches_Link'}, inplace = True)
```
